„ÄêProblem 1] Number of grains of rice on the 100th day

Create a code to calculate how many grains of rice you receive in total on the 100th day, and output the answer. Also, use a line graph to represent the change in the number of rice grains you receive on that day and the total number of rice grains you receive by that day from day 1 to day 100. (The vertical axis is the number of rice and the horizontal axis is the number of days)



```python
rice_grains = 2**99
print("The number of grains of rice on the 100th day is: {}".format(rice_grains))
```

```python
number_rice = 1
days = 100
list_number_rice = [number_rice]
list_total_rice = [number_rice]

for i in range(days - 1):
    number_rice *= 2
    list_number_rice.append(number_rice)
    list_total_rice.append(sum(list_number_rice))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("rice by day")
plt.xlabel("number of days")
plt.ylabel("number of rice")
plt.plot(list_number_rice, label="number of grains")
plt.plot(list_total_rice, label="accumulation of grains", color="blue")
plt.show()
```

„ÄêProblem 2] Number of rice grains outside of the 100th day

```python
def compute_sorori_shinzaemon(day=100):
    """
    A function that verifies the anecdote of rice by Sorori Shinzaemon. Calculate the number of rice grains per day.
    Parameteres
    --------------
    day: int
        Number of days to get rice (default : 100)
    Returns
    --------------
    list_n_grains : list
        List of the number of rice you get one day (the length of the list is equal to day)
    list_total_grains : list
        List of the total number of rice you will receive by a certain day (the length of the list will be equal to day)
    """
    # Write the code here. pass is an instruction that means do nothing.
    
    n_grains = 1
    list_n_grains = [n_grains]
    list_total_grains = [n_grains]
    
    for i in range(days -1):
        n_grains *= 2
        list_n_grains.append(n_grains)
        list_total_grains.append(sum(list_n_grains))
    
    plt.title("rice by day")
    plt.xlabel("number of days")
    plt.ylabel("number of rice")
    plt.plot(list_n_grains, label = 'number of grains', color="red")
    plt.plot(list_total_grains, label = 'accumulation of grains', color="green")
    plt.show()
    
    return list_n_grains, list_total_grains
    
```

```python
list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=100)
```
